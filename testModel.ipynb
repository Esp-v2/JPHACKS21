```python
import os
import keras
from keras.preprocessing.image import ImageDataGenerator
from keras.models import Sequential, Model
from keras.layers import Input, Dense, Dropout, Activation, Flatten,MaxPooling2D,Conv2D
from keras import optimizers
from tensorflow.keras.optimizers import RMSprop
from tensorflow.keras.optimizers import Adam
```

```python
classes = ['neutral', 'happy']
nb_classes = len(classes)
batch_size_for_data_generator = 20

test_dir = '/content/drive/MyDrive/JPHACKS21/test'

# 要調整
img_rows, img_cols = 180, 180
```

```python
test_datagen = ImageDataGenerator(rescale=1.0 / 255)
```

```python
model = keras.models.load_model('/content/drive/MyDrive/JPHACKS21/SmileEstimation_Genki/GENKI-4K/sprit_data/smileEstimation_4000data.hdf5')
```

**test**

```python
test_generator = test_datagen.flow_from_directory(directory=test_dir,
                                                  target_size=(img_rows, img_cols),
                                                  color_mode='rgb',
                                                  classes=classes,
                                                  class_mode='categorical',
                                                  batch_size=batch_size_for_data_generator)

test_loss, test_acc = model.evaluate(test_generator, steps=50)
print('test acc:', test_acc)
```
